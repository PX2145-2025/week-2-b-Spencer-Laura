## More about functions
Sometimes, there can be lots of variables that remain the same when calling a function. If that’s the case you can simplify the syntax a bit. For example, let’s write a function to find how long it takes a mass to hit the ground when dropped (from rest) from a height h. The function will look like:

````python
def T(h, g):
    '''
    Returns fall time [s] of a mass released from a height
    h [m] above the Earth's surface. g in ms^-2.
    '''
    return sqrt(2*h/g)
````

This is called as follows:

````python
height = 1  # metre
print ('Time from', height, 'metre', 'is' , T(height, 9.8),'s')
````

You *could* program this with a hard-coded value for g, to make the function call simpler:

````python
def T(h):
    '''
    Returns fall time [s] of a mass released from a height
    h [m] above the Earth's surface.
    '''
    g=9.81 # in ms^-2
    return sqrt(2*h/g)

height = 1 # metre
print ('Time from', height, 'metre', 'is' , T(height),'s')
````

but then of course it wouldn’t work for Mars’ gravity. A better way is

````python
def T(h, g=9.81):
    '''
    Returns fall time [s] of a mass released from a height
    h [m] above a planet's surface. Default g is 9.81 ms^-2
    '''
    return sqrt(2*h/g)

height = 1  # metre
print('Time from', height, 'metre', 'on Earth is' , T(height),'s')
print('Time from', height, 'metre', 'on Mars is' , T(height, g=3.7),'s')
````

````python
# Exercise 1

# Define a function func1(time, amplitude, frequency, phase) which returns a sine wave (as a function of time) of chosen amplitude, frequency and phase.
def func1(time, amplitude=1, frequency=1, phase=0):
  """
  Returns a sine wave as a function of time with adjustable parameters.

  Args:
    time (array): A NumPy array representing time values.
    amplitude (int or float): The amplitude of the sine wave. Defaults to 1.
    frequency (int or float): The frequency of the sine wave. Defaults to 1.
    phase (int or float): The phase shift of the sine wave. Defaults to 0.

  Returns:
    array: A NumPy array representing the sine wave values.
  """
  return amplitude * sin(2 * pi * frequency * time + phase)

# Generate a time array from 0 to 5 with 500 points
time = linspace(0, 5, 500)

# Create a figure for the subplots
figure(figsize=(15, 5))

# Plot 1: Default values
# Amplitude=1, Frequency=1, Phase=0
subplot(1, 3, 1)
plot(time, func1(time))
title("Default Values\n(Amplitude=1, Freq=1, Phase=0)")
xlabel("Time")
ylabel("Amplitude")
grid(True)

# Plot 2: Increased amplitude
# Amplitude=2, Frequency=1, Phase=0
subplot(1, 3, 2)
plot(time, func1(time, amplitude=2))
title("Increased Amplitude\n(Amplitude=2, Freq=1, Phase=0)")
xlabel("Time")
ylabel("Amplitude")
grid(True)

# Plot 3: Increased frequency
# Amplitude=1, Frequency=2, Phase=0
subplot(1, 3, 3)
plot(time, func1(time, frequency=2))
title("Increased Frequency\n(Amplitude=1, Freq=2, Phase=0)")
xlabel("Time")
ylabel("Amplitude")
grid(True)

tight_layout()
show()

# Plot 4: Adjusted phase
# This plot is separate to show the shift more clearly
figure(figsize=(10, 5))
plot(time, func1(time), label="Phase=0 (Default)")
plot(time, func1(time, phase=pi/2), label="Phase=π/2 (Shifted)")
title("Sine Wave with a Phase Shift")
xlabel("Time")
ylabel("Amplitude")
grid(True)
legend()
show()

savefig('func1.png')



````

Question 1: Proof of altitude formula.

````python
# Question 2
# Write a function to evaluate h

# Constants
G = 6.67430e-11  # Universal gravitational constant (m³ kg⁻¹ s⁻²)
M = 5.972e24     # Mass of Earth (kg)
R = 6.371e6      # Radius of Earth (m)

def calculate_altitude(T):
    """
    Calculates the altitude of a satellite in a circular orbit.

    Args:
      T (float): The orbital period in seconds.

    Returns:
      float: The altitude in kilometers.
    """
    if T <= 0:
        return None

    # Calculate the orbital radius r
    r_cubed = (G * M * T**2) / (4 * pi**2)
    r = r_cubed**(1/3)

    # Calculate the altitude h and convert to kilometers
    h = (r - R) / 1000
    return h
````

````python
# Question 3
# Write a program that requests user imput of T and returns altitude in km

def main():
    """
    Requests user input for orbital period and prints the altitude.
    """
    try:
        # Request orbital period in minutes
        period_minutes = float(input("Enter the orbital period in minutes: "))

        # Convert to seconds
        period_seconds = period_minutes * 60

        # Calculate altitude using the function
        altitude_km = calculate_altitude(period_seconds)

        if altitude_km is not None:
            print(f"The altitude of the satellite is approximately {altitude_km:,.2f} km.")
        else:
            print("Invalid input for orbital period.")

    except ValueError:
        print("Invalid input. Please enter a numerical value.")

if __name__ == "__main__":
    main()
````

**IMPORTANT**: The `linfit` function itself must be self contained. It must only take in the x- and y-data, and return m, dm, c and dc. It must not do anything else, such as print or plot data - that should all be done outside the function.

````python
# Exercise 3

# Write a funtion to perfom a least squares fit of data
%%writefile least_squares_fit.py
"""
least_squares_fit.py

A package for a least squares linear regression.
This contains the 'linfit' function, which calculates the slope,
intercept, and their errors for a straight-line fit to x and y data.
"""

def linfit(x, y):
    """
    Performs a least squares linear fit to data arrays x and y.

    Args:
        x (array): Array of independent variable data.
        y (array): Array of dependent variable data.

    Returns:
        tuple: A tuple containing:
            m (float): The slope of the fitted line.
            dm (float): The error in the slope.
            c (float): The intercept of the fitted line.
            dc (float): The error in the intercept.
    """

    N = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(x**2)
    sum_xy = sum(x * y)

    delta = N * sum_x2 - sum_x**2

    # Calculate slope (m) and intercept (c)
    m = (N * sum_xy - sum_x * sum_y) / delta
    c = (sum_x2 * sum_y - sum_x * sum_xy) / delta

    # Calculate error in slope (dm) and intercept (dc)
    y_fit = m * x + c
    residuals_squared = (y - y_fit)**2
    sy2 = sum(residuals_squared) / (N - 2)
    sy = sqrt(sy2)

    dm = sy * np.sqrt(N / delta)
    dc = sy * np.sqrt(sum_x2 / delta)

    return m, dm, c, dc
````

````python
from least_squares_fit import *

# Generate some test data
# Create a straight line and add some random noise
np.random.seed(42)  # for reproducibility
x_data = np.linspace(0, 10, 20)
true_m = 2.5
true_c = 3.0
noise = np.random.normal(0, 1.5, x_data.size)
y_data = true_m * x_data + true_c + noise

# Perform the linear fit using the imported function
m, dm, c, dc = linfit(x_data, y_data)

# Print the results
print(f"Fitted Slope (m): {m:.4f} ± {dm:.4f}")
print(f"Fitted Intercept (c): {c:.4f} ± {dc:.4f}")

# Plot the data and the fitted line
figure(figsize=(10, 6))
scatter(x_data, y_data, label='Original Data', color='blue', zorder=2)
plot(x_data, m * x_data + c, label=f'Fitted Line: y = {m:.2f}x + {c:.2f}', color='red', linestyle='--', zorder=1)

# Add title, labels, and grid
title('Least Squares Linear Fit')
xlabel('x-axis')
ylabel('y-axis')
grid(True)
legend()
show()
````
